# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load The Dataset

```python
df = pd.read_excel('student_performance_100.xlsx')  
print(df.head())
```

# Basic Data Inspection

```python
print(df.info())
df.describe()
```

```python
df.isnull().sum()
```

# Create Total & Average Score

```python
df['Total Score'] = df['Math Score'] + df['Reading Score'] + df['Writing Score']
df['Average Score'] = df['Total Score'] / 3
```

# Overall Performance Analysis

```python
df[['Math Score', 'Reading Score', 'Writing Score']].mean()
```

# Score Distribution

```python
sns.histplot(df['Average Score'], kde=True)
plt.title("Average Score Distribution")
plt.show()
```

# Gender-Based Analysis

```python
sns.boxplot(x='Gender', y='Average Score', data=df)
plt.title("Gender vs Average Score")
plt.show()
```

# Ethnicity-Based Performance

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Ethnicity', y='Average Score', data=df)
plt.title("Average Score by Ethnicity")
plt.xticks(rotation=45)
plt.show()
```

# Parental Education Impact

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Parental Education', y='Average Score', data=df)
plt.xticks(rotation=45)
plt.title("Parental Education vs Student Performance")
plt.show()
```

# Test Preparation Effect

```python
sns.barplot(x='Test Preparation', y='Average Score', data=df)
plt.title("Test Preparation Impact on Scores")
plt.show()
```

# Subject-wise Comparison

```python
df[['Math Score', 'Reading Score', 'Writing Score']].plot(kind='box')
plt.title("Subject Score Comparison")
plt.show()
```

# Correlation Between Subjects

```python
sns.heatmap(df[['Math Score','Reading Score','Writing Score']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Subject Correlation")
plt.show()
```
